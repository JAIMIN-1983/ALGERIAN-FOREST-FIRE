# One Hot Encoding - Categorial into Numeric

```python
df['Temperature'].unique()
```

```python
df['Rain '].unique()
```

```python
 df['ISI'].unique()
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', ' RH', ' Ws']]=df[['day', 'month', 'year', 'Temperature', ' RH', ' Ws']].astype(int)
```

```python
df['DC']=df['DC'].str.replace('14.6 9', '14.69')
```

```python
df['FWI']=df['FWI'].str.replace('fire',df['FWI'].mode()[0])
```

```python
df[['Rain ','FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']]=df[['Rain ','FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df['Classes  '].unique()
```

```python
df['Classes  ']=df['Classes  '].str.strip()
```

```python
df['Classes  ']=df['Classes  '].str.replace('not fire', '0')
df['Classes  ']=df['Classes  '].str.replace('fire', '1')
```

```python
df['Classes  '][160:170]
```

```python
# df.loc[167,'Classes  ']=df['Classes  '].mode()[0]
```

```python
df['Classes  ']=df['Classes  '].astype(int)
```
